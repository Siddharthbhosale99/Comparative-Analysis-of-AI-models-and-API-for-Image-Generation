```python
import pandas as pd
import markovify
```

```python
inp = pd.read_csv('C:/Users/siddh/Downloads/news-date-text.csv')
```

```python
inp.head(5)
```

```python
# Train text model with markovify to generate prompts (headlines)
text_model = markovify.NewlineText(inp.headline_text, state_size=3)
```

```python
# Create text and save the prompts as a list
prompts_2 = []
num_prompts = 20000
```

```python
for _ in range(num_prompts):
    text = text_model.make_sentence()
    if text is not None: 
        prompts_2.append(text)
```

```python
for i, prompt in enumerate(prompts_2[:20000], 1):
    print(f"{i}. A {prompt}")
```

```python
# Save prompts_2 to a file (optional)
with open('Prompts_Generator_2.csv', 'w') as f:
    for prompt in prompts_2[:20000]:  
        f.write(f"A {prompt}\n")
```
